# Hyperparameter Analysis

This notebook analyses the performance of changing hyperparameters.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Threshold Analysis

```python
combined_threshold_0_3 = pd.read_csv("../logs/combined_threshold_0_3.csv")
combined_threshold_0_5 = pd.read_csv("../logs/combined_threshold_0_5.csv")
combined_threshold_0_7 = pd.read_csv("../logs/combined_threshold_0_7.csv")
combined_threshold_0_9 = pd.read_csv("../logs/combined_threshold_0_9.csv")
```

### Analysis on Training

```python
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.title("Train Accuracy vs Epoch")
plt.plot(combined_threshold_0_3['train_accs'], label="0.3")
plt.plot(combined_threshold_0_5['train_accs'], label="0.5")
plt.plot(combined_threshold_0_7['train_accs'], label="0.7")
plt.plot(combined_threshold_0_9['train_accs'], label="0.9")
plt.xlabel("Epoch")
plt.ylabel("Train Accuracy")
plt.legend()

plt.subplot(1, 4, 2)
plt.title("Train Recall vs Epoch")
plt.plot(combined_threshold_0_3['train_recalls'], label="0.3")
plt.plot(combined_threshold_0_5['train_recalls'], label="0.5")
plt.plot(combined_threshold_0_7['train_recalls'], label="0.7")
plt.plot(combined_threshold_0_9['train_recalls'], label="0.9")
plt.xlabel("Epoch")
plt.ylabel("Train Recall")
plt.legend()

plt.subplot(1, 4, 3)
plt.title("Train Precision vs Epoch")
plt.plot(combined_threshold_0_3['train_precisions'], label="0.3")
plt.plot(combined_threshold_0_5['train_precisions'], label="0.5")
plt.plot(combined_threshold_0_7['train_precisions'], label="0.7")
plt.plot(combined_threshold_0_9['train_precisions'], label="0.9")
plt.xlabel("Epoch")
plt.ylabel("Train Precision")
plt.legend()

plt.subplot(1, 4, 4)
plt.title("Train F1 Score vs Epoch")
plt.plot(combined_threshold_0_3['train_f1_scores'], label="0.3")
plt.plot(combined_threshold_0_5['train_f1_scores'], label="0.5")
plt.plot(combined_threshold_0_7['train_f1_scores'], label="0.7")
plt.plot(combined_threshold_0_9['train_f1_scores'], label="0.9")
plt.xlabel("Epoch")
plt.ylabel("Train F1 Score")
plt.legend()

plt.tight_layout()
plt.show()
```

### Analysis on Testing

```python
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.title("Validation Accuracy vs Epoch")
plt.plot(combined_threshold_0_3['val_accs'], label="0.3")
plt.plot(combined_threshold_0_5['val_accs'], label="0.5")
plt.plot(combined_threshold_0_7['val_accs'], label="0.7")
plt.plot(combined_threshold_0_9['val_accs'], label="0.9")
plt.xlabel("Epoch")
plt.ylabel("Val Accuracy")
plt.legend()

plt.subplot(1, 4, 2)
plt.title("Validation Recall vs Epoch")
plt.plot(combined_threshold_0_3['val_recalls'], label="0.3")
plt.plot(combined_threshold_0_5['val_recalls'], label="0.5")
plt.plot(combined_threshold_0_7['val_recalls'], label="0.7")
plt.plot(combined_threshold_0_9['val_recalls'], label="0.9")
plt.xlabel("Epoch")
plt.ylabel("Val Recall")
plt.legend()

plt.subplot(1, 4, 3)
plt.title("Validation Precision vs Epoch")
plt.plot(combined_threshold_0_3['val_precisions'], label="0.3")
plt.plot(combined_threshold_0_5['val_precisions'], label="0.5")
plt.plot(combined_threshold_0_7['val_precisions'], label="0.7")
plt.plot(combined_threshold_0_9['val_precisions'], label="0.9")
plt.xlabel("Epoch")
plt.ylabel("Val Precision")
plt.legend()

plt.subplot(1, 4, 4)
plt.title("Validation F1 Score vs Epoch")
plt.plot(combined_threshold_0_3['val_f1_scores'], label="0.3")
plt.plot(combined_threshold_0_5['val_f1_scores'], label="0.5")
plt.plot(combined_threshold_0_7['val_f1_scores'], label="0.7")
plt.plot(combined_threshold_0_9['val_f1_scores'], label="0.9")
plt.xlabel("Epoch")
plt.ylabel("Val F1 Score")
plt.legend()

plt.tight_layout()
plt.show()
```
